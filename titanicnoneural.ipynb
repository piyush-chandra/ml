```python
import keras
import pandas as pd
import sklearn as sk
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf 
```

```python
dataset_train=pd.read_csv("t_train.csv")
dataset_test=pd.read_csv("t_test.csv")
dataset_train.shape
```

```python
x_train=dataset_train.drop("Survived",axis=1)
y_train=dataset_train["Survived"]
x_test=dataset_test
```

```python
del dataset_train
del dataset_test
```

```python
x_train=x_train.drop('PassengerId',axis=1)
x_train=x_train.drop('Name',axis=1)
x_train=x_train.drop('Ticket',axis=1)
x_train=x_train.drop('Cabin',axis=1)
```

```python
x_test=x_test.drop('Name',axis=1)
x_test=x_test.drop('PassengerId',axis=1)
x_test=x_test.drop('Ticket',axis=1)
x_test=x_test.drop('Cabin',axis=1)
```

```python
x_train=pd.get_dummies(x_train)
x_test=pd.get_dummies(x_test)
```

```python
x_train.Age.median()
```

```python
x_train.Age.fillna(x_train.Age.mean(), inplace=True)
x_test.Age.fillna(x_test.Age.mean(), inplace=True)
```

```python
x_test.Fare.fillna(method='bfill',inplace=True)
```

```python
from sklearn.svm import SVC
clf=SVC(gamma=0.5,C=1.0e4).fit(x_train, y_train).predict(x_test)
print(clf)
```
